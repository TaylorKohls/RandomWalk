```python
#imports any libraries I might use
import numpy as np
import math
import time
import random
%matplotlib notebook
import matplotlib.pyplot as plt
from numpy.random import poisson
%matplotlib inline
import ipywidgets as widgets
import pygame
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.art3d as art3d
```

```python
#defines the walker simulation as a function of times the simulation was run, steps per simulation,
#and whether the function is to return the total distance of each simulation,
#or the position of each step in every simulation
def WalkerSim(nTimes, nSteps):
    #initializes the x,y distance and the x,y position as well as arrays of times for changes in each as empty arrays
    xDist = []
    yDist = []
    xPos = []
    yPos = []
    simArray1 = []
    simArray2 = []
    #runs through each simulation
    for n in range(nTimes):
        #initializes xd and yd as distance traveled in each step, then appends their sums to xDist and yDist
        xd = np.random.choice([-1,1], nSteps)
        yd = np.random.choice([-1,1], nSteps)
        xDist.append(xd.sum())
        yDist.append(yd.sum())
        #initializes xdist and ydist to 0
        xdist = 0
        ydist = 0
        #runs through each step of the simmulation
        for m in range(nSteps):
            #adds x and y distance traveled in each step to xdist and ydistance
            xdist += np.random.choice([-1,1])
            ydist += np.random.choice([-1,1])
            #appends xdist and ydist to xPos and yPos
            xPos.append(xdist)
            yPos.append(ydist)
            #adds to time variable for individual steps
            #this is to keep all variables the same number of dimensions
            #to avoid weird errors with the 3D graphs where time is a parameter
            simArray1.append(n+m-1)
        #adds to the time variable for total simulations
        #this is for the same purpose for simulations of total distance traveled per simulation
        #as simArray1 is for simulations of individual steps
        simArray2.append(n)
    #returns arrays of each simulation's total x distances, total y distances, with their corresponding 1D time variable array
    #and arrays of each individual x position, y position, and corresponding time variable array
    return xDist, yDist, simArray2, xPos, yPos, simArray1
```

```python
#sets number of simulations and steps per simulation to user input
Sims = int(input())
Steps = int(input())
#calls the walker simulation function with the user input as parameters
#then prints the results
distanceArray = WalkerSim(Sims, Steps)
print(distanceArray)
```

```python
#calls the function again with the above imputs and specifies to model individual position
#then uses ax.plot to ctreate a 3D plot
xs = distanceArray[3]
ys = distanceArray[4]
zs = distanceArray[5]
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(xs, ys, zs) 
plt.title('Line Graph')
```

```python
#uses ax.plot to create a 3D bar graph
fig = plt.figure()
ax = plt.axes(projection="3d")

#
num_bars = Sims
x_pos = distanceArray[0]
y_pos = distanceArray[1]
z_pos = [0] * num_bars
x_size = np.ones(num_bars)
y_size = np.ones(num_bars)
z_size = distanceArray[2]

#
ax.bar3d(x_pos, y_pos, z_pos, x_size, y_size, z_size, color='aqua')
plt.show()
```

```python
#uses ax.plot to create a 3D stem graph
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')

N = 100
x = distanceArray[0]
y = distanceArray[1]
z = distanceArray[2]
for xi, yi, zi in zip(x, y, z):        
    line=art3d.Line3D(*zip((xi, yi, 0), (xi, yi, zi)), marker='o', markevery=(1, 1))
    ax.add_line(line)
ax.set_xlim3d(-20, 20)
ax.set_ylim3d(-20, 20)
ax.set_zlim3d(0, N)    
plt.show()
```
